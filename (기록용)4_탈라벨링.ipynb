## 얼굴사진, 탈과 거리

```python
from pandas import read_csv
df = read_csv("/content/drive/MyDrive/matlab 공모전/csv(표)/Standard_distance(수정후).csv")
df
#import pandas as pd
#sim = pd.read_csv("/content/drive/MyDrive/matlab 공모전/csv(표)/use_front_sim.csv")
```

```python
df_val = df.drop(df.columns[[0]], axis=1)
simval = df_val.values
print(simval)
```

## 가장 작은 거리 1번째, 2번째 탈

```python
#사람 사진 수(df의 row 수)
ren = 612
```

```python
import numpy as np
first = np.ones(ren, int)
for i in range(ren):
  first[i] = np.argmin(simval[i])

first
```

```python
second = np.ones(ren, int)
for i in range(ren):
  sorted_dif_idx = np.argsort(simval[i])
  second[i] = sorted_dif_idx[1]
second
```

```python
between = np.ones(ren)
near = np.zeros(ren, int)
for k in range(ren):
  idx1 = first[k]
  idx2 = second[k]
  between[k] = simval[k][idx2] - simval[k][idx1]
  if between[k] < 0.1:
    near[k] = 7
```
